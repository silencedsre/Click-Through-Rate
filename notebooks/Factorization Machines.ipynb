```python
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=10000,n_features=100, n_clusters_per_class=1)
```

```python
data = pd.DataFrame(data = X)
data
```

```python
data = data.to_dict('records')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.1, random_state=42)
```

```python
v = DictVectorizer()
X_train = v.fit_transform(X_train)
X_test = v.transform(X_test)
fm = pylibfm.FM(num_factors=50, num_iter=100, verbose=True, task="classification", initial_learning_rate=0.0001, learning_rate_schedule="optimal")
```

```python
fm.fit(X_train,y_train)
```

```python
# Evaluate
from sklearn.metrics import log_loss
loss = log_loss(y_test,fm.predict(X_test))
loss
```
