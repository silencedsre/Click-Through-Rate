```python
import pandas as pd
from pyfm import pylibfm
from sklearn.feature_extraction import DictVectorizer
from sklearn.datasets import make_classification
```

```python
#making imbalanced dataset with 0 more than 1
X, y = make_classification(n_samples=10000,n_features=100, n_clusters_per_class=1, weights=[0.8])
```

```python
data = pd.DataFrame(data = X)
data
```

```python
dict_data = data.to_dict('records')
#representating features in dictionary
dict_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dict_data, y, test_size=0.2, random_state=42)
```

```python
v = DictVectorizer()
X_train = v.fit_transform(X_train)
X_test = v.transform(X_test)
fm = pylibfm.FM(num_factors=50, num_iter=10, verbose=True, task="classification", initial_learning_rate=0.0001, learning_rate_schedule="optimal")
```

```python
fm.fit(X_train,y_train)
```

```python
# Evaluate
from sklearn.metrics import log_loss
loss = log_loss(y_test,fm.predict(X_test))
loss
```

```python
pred = fm.predict(X_test)
```

```python
data_test = pd.DataFrame(y_test)
```

```python
data_test.columns = ['actual']
```

```python
data_test['pred'] = pred
```

```python
def actual_pred(value):
    if value<0.5:
        return 0
    else:
        return 1
```

```python
new_pred = []
for val in pred:
    temp = actual_pred(val)
    new_pred.append(temp)
data_test['new_pred'] = new_pred
```
