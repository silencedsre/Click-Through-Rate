```python
from pyfm import pylibfm
import pandas as pd
# Transforms lists of feature-value mappings to vectors.
from sklearn.feature_extraction import DictVectorizer
# Generate a random n-class classification problem
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=10000, n_features=100, n_clusters_per_class=1, weights=[0.8])
```

```python
data = pd.DataFrame(data=X)
data
```

```python
data_dict = data.to_dict('records')
data_dict
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_dict, y, test_size=.2, random_state=42)
```

```python
v = DictVectorizer()
X_train = v.fit_transform(X_train)
X_test = v.transform(X_test)
fm = pylibfm.FM(num_factors=10, num_iter=10, verbose=True, task="classification", initial_learning_rate=0.0001, learning_rate_schedule="optimal")
```

```python
fm.fit(X_train, y_train)
```

```python
from sklearn.metrics import log_loss
loss = log_loss(y_test,fm.predict(X_test))
loss
```

```python
pred = fm.predict(X_test)
```

```python
data_test = pd.DataFrame(y_test)
```

```python
data_test.columns = ['actual']
```

```python
data_test['pred'] = pred
```

```python
data_test
```

```python
temp_df = data_test.describe()
temp_df
```

```python
diff = temp_df.iloc[7]['pred'] - temp_df.iloc[3]['pred']
diff/2
```

```python
def actual_pred(value):
    if value<0.5:
        return 0
    else:
        return 1
```

```python
new_pred = []
for val in pred:
    temp = actual_pred(val)
    new_pred.append(temp)
data_test['new_pred'] = new_pred
```

```python
data_test = data_test.drop('pred', axis=1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, new_pred))
```
